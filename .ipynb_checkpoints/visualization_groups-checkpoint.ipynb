```python
def get_model_name(model_type):
    match model_type:
        case 1:
            return 'LegNet'
        case 2:
            return 'LegNet&PWMs'
        case 3:
            return 'PWMNet'
        case 4:
            return 'PWM'
        case 5:
            return 'ArChIPelago'
def get_model_color(model_type):
    match model_type:
        case 1:
            return 'crimson'
        case 2:
            return 'green'
        case 3:
            return 'orange'
        case 4:
            return 'mediumorchid'
        case 5:
            return 'darkblue'
        case _:
            if isinstance(model_type, int):
                raise ValueError('No such key')
    return get_model_color(get_model_name(model_type))
def get_mode_marker(mode):
    match mode:
        case 'Training':
            return '^'
        case 'Validation':
            return 'X'
        case 'Testing':
            return 'o'
model_palette = {get_model_name(x):get_model_color(x) for x in range(1, 6)}
marker_collection = {x:get_mode_marker(x) for x in ('Training', 'Validation', 'Testing')}
experiments = ('CHS', 'GHTS')
size=150
alpha=0.6
```

```python
preds = ('forw_pred', 'avg_pred', 'rev_pred')
preds_names = {'forw_pred': 'forward', 'rev_pred': 'reverse', 'avg_pred': 'averaged'}

def transform_metrics_df(df):
    left_df = df[['val_loss', 'val_auroc', 'epoch', 'step']].dropna()
    right_df = df[['epoch', 'step', 'train_loss']].dropna()
    return pd.merge(left_df, right_df, on=['epoch', 'step'])

def transform_preds_df(df):
    df['avg_pred'] = (df['forw_pred'] + df['rev_pred']) / 2
    return df
```

```python
def get_proper_validation_metric(df, name):
    # return df.loc[df['val_loss'].idxmin()][name]
    return df.loc[df[name].idxmax()][name]
```

```python
def group_paths(to_group, path_depth, convert_key=int):
    grouped = defaultdict(dict)
    for entry in to_group:
        name_full = entry.parents[path_depth].name
        name = name_full[:name_full.rindex('-')]
        key = convert_key(name_full[name_full.rindex('-')+1:])
        grouped[name][key] = entry
    return grouped

def read_groups(paths, sep, transform=None):
    read = defaultdict(dict)
    for name, value in paths.items():
        for key, path in value.items():
            df = pd.read_csv(path, sep=sep)
            if transform:
                df = transform(df)
            read[name][key] = df
    return read
```

```python
def plot_scatterplot_metrics(df, metric_name, ylim=0.0, figsize=None, title=None):
    sns.set_style('whitegrid')
    if figsize is None:
        figsize = [x / 2 for x in (20, 15)]
    if title is None:
        title = f'{metric_name} values'
    fig, axs = plt.subplots(1, 2, figsize=figsize, layout='constrained',
                                  sharex=False, sharey=True)
    fig.suptitle(title)
    for exp1, exp2, ax in zip(experiments, reversed(experiments), axs.flat):
        plot_values_exp = df.loc[df['Experiment'] == exp1]

        g = sns.scatterplot(data=plot_values_exp, 
                            x='TF', 
                            y=metric_name, 
                            hue='Model type', 
                            s=size, 
                            alpha=alpha, 
                            palette=model_palette, 
                            markers=marker_collection, 
                            ax=ax)
        sns.move_legend(g, "upper left", bbox_to_anchor=(1, 1))
        g.set(ylim=(ylim, 1))
        ax.axhline(y=0.5, ls=':', c='grey')
        g.set(title=f'{exp1}/{exp2}')
        g.yaxis.set_major_locator(ticker.MultipleLocator(0.1))
        ax.get_legend().set_visible(False)
    handles, labels = ax.get_legend_handles_labels()
    fig.legend(handles, labels, loc='lower right', bbox_to_anchor=(1,1))
```

```python
def get_dataset_statistics():
    replace_names = {
        'positives.bed': 'positives',
        'foreigns.bed': 'foreigns',
        'random_addshift.bed': 'random',
        'shades_addshift.bed': 'shades'
    }
    path_to_ds_statistics = path_to_base / 'MEX/Archipelago_data'
    print(path_to_ds_statistics)
    ds_stat_paths = path_to_ds_statistics.rglob('dataset_statistics_*.txt')
    ds_stat = [pd.read_csv(x, sep=',', header=None) for x in ds_stat_paths]
    ds_stat = pd.concat(ds_stat)
    ds_stat[0] = ds_stat[0].str.strip()
    ds_stat[[0, 1]] = ds_stat[0].str.split(' ', n=1, expand=True)
    ds_stat[list(range(1, 5))] = ds_stat[1].str.rsplit('/', n=4, expand=True).drop(0, axis=1)
    ds_stat.columns = ['Count', 'Experiment', 'Split', 'TF', 'Seq type']
    ds_stat[['Count']] = ds_stat[['Count']].astype(int)
    return ds_stat.replace(replace_names).dropna()

get_dataset_statistics()
```

```python
def get_dataset_statistics_():
    replace_names = {
        'positives.bed': 'positives',
        'foreigns.bed': 'foreigns',
        'random_addshift.bed': 'random',
        'shades_addshift.bed': 'shades'
    }
    path_to_ds_statistics = path_to_base / 'MEX/Archipelago_data'
    print(path_to_ds_statistics)
    ds_stat_paths = path_to_ds_statistics.rglob('dataset_statistics_*.txt')
    ds_dfs = []
    for ds_path in ds_stat_paths:
        ds_df = pd.DataFrame(columns=['Experiment', 'Split', 'Seq type', 'TF', 'Count'])
        with open(ds_path, 'r') as ds_file:
            for line in ds_file.readlines():
                clear_line = line.lstrip()
                count, path = clear_line.split()
                path_split = list(reversed(path.split('/')))
                if len(path_split) < 4:
                    continue
                seq_type_bed, tf, split, exp, *_ = path_split
                ds_df.loc[len(ds_df)] = [exp, split, replace_names[seq_type_bed], tf, int(count)]
        ds_dfs.append(ds_df)
    return pd.concat(ds_dfs)
    
get_dataset_statistics_()
```

```python
def extend_row(row):
    row = list(row)
    return row + [row[0]]

def plot_radarplot_metrics(df, metric_name, ylim=0.0, figsize=None, title=None, set_y_min_max=False, exclude_models=None):
    sns.set_style('whitegrid')
    setRCParams()
    # rcParams["axes.axisbelow"] = True
    if exclude_models is None:
        exclude_models = []
    if figsize is None:
        figsize = [x / 1.5 for x in (20, 10)]
    if title is None:
        title = f'{metric_name} values, random  {negatives_to} negatives'
    fig, axs = plt.subplots(1, 2, figsize=figsize, layout='constrained',
                                  sharex=False, sharey=False, subplot_kw=dict(projection="polar"))
    fig.suptitle(title, size=15, fontweight='medium')
    
    plot_vals = df.loc[~df['Model type'].isin(exclude_models)]
    out_vals = []
    for exp1, exp2, ax in zip(experiments, reversed(experiments), axs.flat):
        plot_vals_exp = plot_vals.loc[plot_vals['Experiment'] == exp1]
        pivoted = plot_vals_exp.pivot(index='Model type', columns='TF', values=metric_name)

        label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(pivoted.columns)+1)
        dct = {}
        for idx, row in pivoted.iterrows():
            ax.plot(label_loc, extend_row(row), label=idx, color=model_palette[idx], linewidth=2)
            if idx == 'LegNet' or idx == 'LegNet&PWMs':
                dct[idx] = row
                
        ax.set_title(f'{get_exp_name(exp1)}  {get_exp_name(exp2)}', fontweight='medium')
        ax.set_zorder(3)
        lines, labels = ax.set_thetagrids(np.degrees(label_loc[:-1]), labels=pivoted.columns)
        for label, text in zip(labels, pivoted.columns):
            best_model = pivoted[text].idxmax()
            label.set(color=model_palette[best_model])
            label.set(zorder=100000)
            label.set_fontweight('semibold')
        if set_y_min_max:
            delta = 0.01
            min_y = max(0, np.nanmin(pivoted.to_numpy()) - delta)
            max_y = min(1, np.nanmax(pivoted.to_numpy()) + delta)
            ax.set_rlim(min_y, max_y)
            ax.set_rticks(np.arange(np.floor(min_y * 10) / 10, np.ceil(max_y * 10) / 10, 0.1))
        else:
            ax.set_ylim(ylim, 1)
            ax.set_rticks(np.arange(0, 1.1, 0.1))
        try:
            out_vals.append(dct['LegNet&PWMs'] - dct['LegNet'])
        except: 
            print('No model selected')
    handles, labels = ax.get_legend_handles_labels()
    fig.legend(handles, labels, loc='lower right', bbox_to_anchor=(1,0.95), title='Model')
    minmaxed = 'minmaxed' if set_y_min_max else ''
    excl = '+'.join(exclude_models)
    plt.savefig(f'output/radarplot_metrics_{metric_name}_{negatives_to}_{minmaxed}_{excl}.png'.replace(' ', '_'),bbox_inches='tight', dpi=300)
    return out_vals
```

```python
def normalize_list(values, log=False):
    if not log:
        max_value = max(values)
        return [x / max_value for x in values]
    if log:
        log_vals = np.log10(values)
        max_value = np.max(log_vals)
        return log_vals / max_value
```

```python
def set_min_max(values, min_val, max_val):
    values = np.array(values)
    was_min = np.min(values)
    values = (values - was_min) / (np.max(values) - was_min)
    return values * (max_val - min_val) + min_val
```

```python
def plot_radarplot_ds(df, negatives=None, ylim=0.0, figsize=None, title=None, set_y_min_max=False, enable_thetagrids=True):
    seq_type_colors = {
        'positives': 'tab:blue',
        'foreigns': 'tab:orange',
        'random': 'tab:red',
        'shades': 'tab:green'
    }
    sns.set_style('whitegrid')
    setRCParams()
    if figsize is None:
        figsize = [x / 1.5 for x in (20, 10)]
    if title is None:
        title = 'Dataset sizes (peaks count)'
    fig, axs = plt.subplots(1, 2, figsize=figsize, layout='constrained',
                                  sharex=False, sharey=False, subplot_kw=dict(projection="polar"))
    fig.suptitle(title) 
    plot_vals = df if negatives is None else df.loc[df['Seq type'].isin(['positives', negatives])]
    amnts = []
    for exp1, exp2, ax in zip(experiments, reversed(experiments), axs.flat):
        # ax.set_axisbelow(False)  
        plot_vals_exp = plot_vals.loc[plot_vals['Experiment'] == exp1]
        # plot_vals_exp = plot_vals.loc[plot_vals['Seq type'] == 'positive']
        # plot_vals_exp = plot_vals.loc[plot_vals['Split'] == 'Train']
        plot_vals_exp = plot_vals_exp.loc[plot_vals_exp['TF'].isin(tfs_by_exp[exp1])]
        pivoted = plot_vals_exp.pivot(index=['Seq type', 'Split'], columns='TF', values='Count')
        label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(pivoted.columns)+1)
        
        for (seq_type, split), row in pivoted.iterrows():
            c = 'tab:blue' if seq_type == 'positives' else 'tab:red'
            l = '-' if split == 'Train' else ':'
            if seq_type != 'positives' or split != 'Train':
                continue
            amounts = row
            ax.scatter(label_loc[:-1], [0.75] * 10, label=f'{seq_type.capitalize()}, {split}', color='#4984ec', s=set_min_max(np.log10(row)**2, 50, 3000))
        # ax.scatter(label_loc[:-1], [0.75] * 10, s=set_min_max(np.log10([1e2, 1e2, 1e2,/ 1e2, 1e2, 1e2, 1e3, 1e4, 1e5, 1e6]), 50, 3000), color='r')
        ax.set_title(f'{get_exp_name(exp1)}', fontweight='semibold')
        ax.yaxis.set_ticklabels([])
        if enable_thetagrids:
            lines, labels = ax.set_thetagrids(np.degrees(label_loc[:-1]), labels=[f'{x}\n{y}' for x,y in zip(pivoted.columns, amounts)])
            for label in labels:
                label.set(zorder=100000)
                label.set_fontweight('semibold')
        else:
            lines, labels = ax.set_thetagrids(np.degrees(label_loc[:-1]), labels=['' for x in pivoted.columns])
            ax.set_rticks([])
        ax.set_rlim(0, 1)
        amnts.append(amounts)

    
    handles, labels = ax.get_legend_handles_labels()
    neg_used = 'negatives_all' if negatives is None else f'positives_{negatives}' 
    opq = '' if enable_thetagrids else f'_opaque' 
    plt.savefig(f'output/radarplot_ds_{neg_used}{opq}.png',bbox_inches='tight', dpi=300)
    
    return amnts

ds_stats = get_dataset_statistics_().drop_duplicates()
ds_sizes = plot_radarplot_ds(ds_stats)
_ = plot_radarplot_ds(ds_stats, enable_thetagrids=False)
```

```python
import matplotlib.colors as mc
import colorsys
from matplotlib.colors import rgb_to_hsv, hsv_to_rgb, to_rgb

def add_opacity(to, alpha):
    return list(to) + [alpha, ]

def set_opacity(of, alpha):
    of = list(of)
    of[3] = alpha
    return of

def lighten_color(color, amount=0.5):  
    # --------------------- SOURCE: @IanHincks ---------------------
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = colorsys.rgb_to_hls(*mc.to_rgb(c))
    return add_opacity(colorsys.hls_to_rgb(c[0], 1 - amount * (1 - c[1]), c[2]), color[3])

def enlighten(color, factor=0.5):
    h, s, v = rgb_to_hsv(to_rgb(color))
    return add_opacity(hsv_to_rgb((h, s, 1 - factor * (1 - v))), color[3])

def endarken(color, factor=0.5):
    h, s, v = rgb_to_hsv(to_rgb(color))
    return add_opacity(hsv_to_rgb((h, s, factor * v)), color[3]) 


def draw_boxplots(exclude_models=None): 
    fig, axes = plt.subplots(2, 2, figsize=(11, 13), sharex=False, sharey=False)
    title = f'Training random  Testing {negatives_to} negatives'
    fig.suptitle(title, size=15, fontweight='medium')
    setRCParams()
    sns.set_style('white')
    flierprops = dict(marker='o', markersize=3, alpha=0)
    pr_baseline = 1 / 3 if negatives_to == 'shades' else 1 / 101
    metrics = ['auROC', 'auROC', 'auPRC', 'auPRC']
    baselines = [0.5, 0.5, pr_baseline, pr_baseline]
    exps = ['GHTS', 'CHS'] * 2
    if exclude_models is None:
        exclude_models = []
    for ax, metric, baseline, exp, exp2 in zip(axes.flat, metrics, baselines, exps, reversed(exps)):
        print(ax, metric, baseline, exp)
        plot_values = plot_values_all.loc[plot_values_all['Experiment'] == exp]
        plot_values = plot_values.loc[~plot_values['Model type'].isin(exclude_models)]
        sns.boxplot(data=plot_values, x="Model type", y=metric, palette=model_palette, saturation=1, flierprops=flierprops, ax=ax)
        sns.swarmplot(data=plot_values, x="Model type", y=metric, hue="Model type", palette=model_palette, ax=ax, alpha=0.5)
        
        ax.get_legend().set_visible(False)
        ax.set_title(f'{get_exp_name(exp)}  {get_exp_name(exp2)}', fontweight='medium', fontsize=16)
        box_container = ax.patches if len(ax.patches) > 0 else ax.artists
        lines_per_boxplot = len(ax.lines) // len(box_container)
        for i, (box, xtick) in enumerate(zip(box_container, ax.get_xticklabels())):
            xtick.set_color(endarken(box.get_facecolor()))
            xtick.set_fontweight('semibold')
            #xtick.set_fontsize(9.7)
            color = enlighten(box.get_facecolor())
            box.set_color(color)
            box.set_facecolor(set_opacity(color, 0.3))

            # Each box has 6 associated Line2D objects (to make the whiskers, fliers, etc.)
            # Loop over them here, and use the same colour as above
            # for j in range(i*6,i*6+6):
            #     line = ax.lines[j]
            #     line.set_color(col)
            #     line.set_mfc(col)
            #     line.set_mec(col)
            #     line.set_linewidth(1)   # ADDITIONAL ADJUSTMENT
            
            for lin in ax.lines[i * lines_per_boxplot: (i + 1) * lines_per_boxplot]:
                if lin.get_label() != 'median':
                    lin.set_color(color)
                    lin.set_markerfacecolor(color)
                    lin.set_markeredgecolor(color)
        
        ax.axline([0, baseline], [1, baseline], linewidth=1, color='black', linestyle=':', zorder=0)
        ax.set_ylim(0, 1)
        for tick in ax.get_xticklabels():
            tick.set_rotation(45)
        ax.set_xlabel('')
        # b.axes.set_title("Title",fontsize=50)
        # b.set_xlabel("X Label",fontsize=30)
        # b.set_ylabel("Y Label",fontsize=20)
        # b.tick_params(labelsize=5)

        
    excl = '+'.join(exclude_models)
    plt.tight_layout()
    plt.savefig(f'output/boxplot_{negatives_to}_{excl}.png',bbox_inches='tight', dpi=300)

draw_boxplots(['PWMNet'])
```
